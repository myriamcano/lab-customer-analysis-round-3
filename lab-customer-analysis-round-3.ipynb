```python
import pandas as pd

data=pd.read_csv('finaldata')
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.countplot(x='gender', data=data) 
plt.xlabel('gender')
plt.ylabel('count')
plt.title('Gender Breakdown')
```

```python
sns.barplot(x='state', y='income', data=data)
plt.xlabel('State')
plt.ylabel('Income')
plt.title('Income by State')
plt.show()
```

```python
sns.barplot(x='policy_type', y='total_claim_amount', data=data)
plt.xlabel('Policy Type')
plt.ylabel('Total Claim Amount')

plt.show()
```

```python
sns.countplot(x='vehicle_class', hue='education', data=data)
plt.xlabel('Vehicle Class')
plt.ylabel('Count')
plt.title('Education by Vehicle Class')
plt.legend(title='Education', loc='upper right', bbox_to_anchor=(1.2, 1.0))  # Display legend
plt.show()
```
